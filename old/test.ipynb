```python
import pandas as pd
import io
from typing import Tuple, Optional

def create_virtual_card_lookup(csv_data: str) -> dict:
    """
    Reads the CSV data and creates a lookup dictionary for virtual card ranges.
    The structure is {bin: [(start_range, end_range), ...]}
    """
    df = pd.read_csv(io.StringIO(csv_data))
    
    lookup_dict = {}
    
    # Group by BIN to process all ranges for a given BIN together
    for bin_code, group in df.groupby('BIN'):
        if bin_code not in lookup_dict:
            lookup_dict[str(bin_code)] = []
        
        for _, row in group.iterrows():
            # Pad ranges to ensure they are 9 digits for consistent comparison
            start_range = str(row['RANGE INICIAL']).zfill(9)
            end_range = str(row['RANGE FINAL']).zfill(9)
            lookup_dict[str(bin_code)].append((start_range, end_range))
            
    return lookup_dict

def is_virtual_card(card_number: str, lookup_dict: dict) -> Tuple[bool, Optional[str]]:
    """
    Checks if a given card number is a virtual card based on the lookup dictionary.

    Args:
        card_number: The card number as a string.
        lookup_dict: The dictionary generated by create_virtual_card_lookup.

    Returns:
        A tuple containing:
        - A boolean (True if the card is virtual, False otherwise).
        - A string describing the matched rule, or None if no match was found.
    """
    # Ensure the card number is a string
    card_number = str(card_number)

    # Basic validation for card number length
    if len(card_number) < 15: # A typical minimum card length
        return (False, None)

    # Extract the first 6 digits for the BIN and the last 9 for the range
    card_bin = card_number[:6]
    card_range_part = card_number[-9:]

    # Check if the BIN exists in our lookup dictionary
    if card_bin in lookup_dict:
        # Iterate through all the ranges associated with this BIN
        for start_range, end_range in lookup_dict[card_bin]:
            # Check if the card's range part falls within any of the virtual ranges
            if start_range <= card_range_part <= end_range:
                rule_description = f"Matched BIN {card_bin} with range {start_range}-{end_range}"
                return (True, rule_description)
    
    # If the BIN is not found or no range matches, it's not a virtual card
    return (False, None)

def check_and_print(card_id, card_num, rules):
    """Helper function to print the check results."""
    is_virtual, rule = is_virtual_card(card_num, rules)
    print(f"Is {card_id} ({card_num}) virtual? {is_virtual}")
    if is_virtual:
        print(f"  -> Rule: {rule}\n")
    else:
        print("") # Just add a newline for spacing

df = pd.read_excel('bin_range_santander.xlsx')
virtual_card_rules = create_virtual_card_lookup(df.to_csv(index=False))

df
```

```python
def generate_corrected_sql_case_statement(csv_data: str) -> str:
    """
    Reads CSV data and generates a corrected, well-formatted SQL CASE statement.
    """
    df = pd.read_excel(csv_data)

    # Group the dataframe by the 'BIN' column
    grouped = df.groupby('BIN')

    all_bin_conditions = []

    # Iterate over each group (each unique BIN)
    for bin_code, group in grouped:
        range_conditions = []
        # For each BIN, iterate over its associated ranges
        for index, row in group.iterrows():
            # Get the start and end ranges directly. No padding is needed.
            range_start = row['RANGE INICIAL']
            range_end = row['RANGE FINAL']
            
            # Create a numeric comparison by casting the card's substring to a number.
            # Using BIGINT for broad compatibility.
            range_conditions.append(f"(CAST(SUBSTR(a.card, 7, 9) AS BIGINT) BETWEEN {range_start} AND {range_end})")
        
        # Join the multiple range conditions for a single BIN with 'OR' and add indentation
        bin_specific_condition = f" OR \n                ".join(range_conditions)
        
        # Create the full condition for this BIN, combining the bin_code check (first 6 digits) with its range checks
        full_bin_condition = f"(SUBSTR(a.card, 1, 6) = '{bin_code}' AND (\n                {bin_specific_condition}\n            ))"
        all_bin_conditions.append(full_bin_condition)

    # Join all the master conditions for each BIN with 'OR' and add indentation
    final_sql_condition = "\n          OR ".join(all_bin_conditions)

    # Construct the final, complete CASE statement with proper formatting
    sql_case_statement = f"""CASE 
    WHEN {final_sql_condition}
    THEN 1 
    ELSE 0 
END as flag_cartao_virtual"""

    return sql_case_statement

print(generate_corrected_sql_case_statement('bin_range_santander.xlsx'))
```
